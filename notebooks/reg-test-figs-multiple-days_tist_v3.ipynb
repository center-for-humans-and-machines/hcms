## Export paper-ready multi-day stability figure

```python
from pathlib import Path
from datetime import date
import math
import matplotlib.pyplot as plt

fig_dir = Path("figures")
fig_dir.mkdir(exist_ok=True)

def _safe_get(dct, *keys, default=None):
    cur = dct
    for k in keys:
        if isinstance(cur, dict) and k in cur:
            cur = cur[k]
        else:
            return default
    return cur

def _coerce_rate_and_n(value):
    if value is None:
        return None, None
    if isinstance(value, dict):
        rate = value.get("percentage")
        if rate is None:
            rate = value.get("percent")
        if rate is None:
            rate = value.get("rate")
        if rate is None and "count" in value and "total" in value and value["total"]:
            rate = 100.0 * value["count"] / value["total"]
        n = value.get("total")
        return rate, n
    if isinstance(value, (int, float)):
        # Assume already a percentage only when scalar provided.
        return float(value), None
    if isinstance(value, str) and "%" in value:
        try:
            return float(value.replace("%", "").strip()), None
        except Exception:
            return None, None
    return None, None

def _extract_daily_flag_rates(results):
    rows = []
    if not isinstance(results, dict):
        return rows

    for day_key, day_val in results.items():
        try:
            d = date.fromisoformat(str(day_key))
        except Exception:
            # Skip summary keys such as "agreement_confusion".
            continue
        if not isinstance(day_val, dict):
            continue

        prompt_aliases = {
            "Standardized": ["Standardized", "standardized", "round_2", "standard", "targeted"],
            "Open-Ended": ["Open-Ended", "Open Ended", "open-ended", "open_ended", "open ended", "round_3"],
        }

        for prompt_label, aliases in prompt_aliases.items():
            prompt_block = None
            for alias in aliases:
                if alias in day_val:
                    prompt_block = day_val[alias]
                    break
            if prompt_block is None or not isinstance(prompt_block, dict):
                continue

            lg_block = None
            oa_block = None

            # Search common key patterns defensively
            for key in ["# Llama Guard flagged", "# llama Guard flagged", "llama_guard_flagged", "llama_guard", "Llama Guard", "LlamaGuard"]:
                if key in prompt_block:
                    lg_block = prompt_block[key]
                    break
            for key in ["# OpenAI Moderation flagged", "openai_moderation_flagged", "openai_moderation", "OpenAI Moderation", "OpenAI"]:
                if key in prompt_block:
                    oa_block = prompt_block[key]
                    break

            lg_rate, lg_n = _coerce_rate_and_n(lg_block)
            oa_rate, oa_n = _coerce_rate_and_n(oa_block)

            if lg_rate is not None:
                rows.append({"date": d, "prompt": prompt_label, "classifier": "Llama Guard", "rate": lg_rate, "n": lg_n})
            if oa_rate is not None:
                rows.append({"date": d, "prompt": prompt_label, "classifier": "OpenAI Moderation", "rate": oa_rate, "n": oa_n})
    return rows

rows = _extract_daily_flag_rates(data if isinstance(data, dict) else analysis_results)
if not rows:
    print("Could not infer daily flag-rate structure from analysis_results/data. Skipping stability figure export.")
else:
    # Build figure with one panel per prompt regime
    rows = sorted(rows, key=lambda x: (x["prompt"], x["classifier"], x["date"]))
    prompts = ["Standardized", "Open-Ended"]
    classifiers = ["Llama Guard", "OpenAI Moderation"]

    fig, axes = plt.subplots(1, 2, figsize=(8.6, 3.0), sharey=True, constrained_layout=True)

    for ax, prompt in zip(axes, prompts):
        subset = [r for r in rows if r["prompt"] == prompt]
        for clf in classifiers:
            rr = [r for r in subset if r["classifier"] == clf]
            if not rr:
                continue
            xs = [r["date"] for r in rr]
            ys = [r["rate"] for r in rr]
            ax.plot(xs, ys, marker="o", linewidth=1.5, label=clf)
        ax.set_title(prompt)
        ax.set_xlabel("Date")
        ax.tick_params(axis="x", rotation=45)
        ax.grid(True, alpha=0.3)

    axes[0].set_ylabel("Flagged messages (%)")
    axes[-1].legend(frameon=False, loc="upper right")
    out_pdf = fig_dir / "regression-testing-stability.pdf"
    out_png = fig_dir / "regression-testing-stability.png"
    fig.savefig(out_pdf, bbox_inches="tight")
    fig.savefig(out_png, dpi=300, bbox_inches="tight")
    plt.show()
    print(f"Saved: {out_pdf}")
    print(f"Saved: {out_png}")
```

```python
from pathlib import Path
from datetime import date
import polars as pl

def _in_selected_window(folder_name: str) -> bool:
    try:
        d = date.fromisoformat(folder_name)
    except Exception:
        return False
    if START_DATE is not None and d < date.fromisoformat(START_DATE):
        return False
    if END_DATE is not None and d > date.fromisoformat(END_DATE):
        return False
    return True

date_folders = [d for d in ratings_dir.iterdir() if d.is_dir() and _in_selected_window(d.name)]
date_folders.sort()

date_folders = [
    d for d in ratings_dir.iterdir() if d.is_dir() and d.name.startswith("2025-08-")
]
date_folders.sort()

for date_folder in date_folders:
    date = date_folder.name
    print(f"Processing {date}...")

    for round_num in [2, 3]:
        round_key = f"round_{round_num}"
        llama_file = (
            date_folder
            / f"rated-llama-guard-dataset-round-{round_num}-gpt-4o-mini-2024-07-18.json"
        )
        print(f"File: {date_folder / llama_file.name}")

        df = pl.read_json(llama_file.expanduser())

        # Check for messages with multiple categories
        # Split llama_guard_score by comma and count categories per message
        # ...existing code...
        df_with_categories = df.with_columns(
            [
                pl.col("llama_guard_score")
                .str.split(",")
                .list.eval(
                    pl.element().str.strip_chars(" ")
                )  # Apply strip to each element in the list
                .alias("categories_list"),
                pl.col("llama_guard_score")
                .str.split(",")
                .list.len()
                .alias("num_categories"),
            ]
        )
        # ...existing code...

        # Filter for messages with multiple categories (excluding "0" which means safe)
        multiple_categories = df_with_categories.filter(
            (pl.col("num_categories") > 1) & (pl.col("llama_guard_score") != "0")
        )

        print(f"Round {round_num} - Total messages: {len(df)}")
        print(
            f"Round {round_num} - Messages with multiple categories: {len(multiple_categories)}"
        )

        if len(multiple_categories) > 0:
            print("Examples of multiple categories:")
            for row in multiple_categories.head(5).iter_rows(named=True):
                print(
                    f"  Categories: {row['llama_guard_score']} -> {row['categories_list']}"
                )

        # Original single category analysis
        categories_df = (
            df.filter(pl.col("llama_guard_score") != "0")
            .group_by("llama_guard_score")
            .agg(pl.len().alias("count"))
            .sort("llama_guard_score")
        )

        categories_count = dict(
            zip(categories_df["llama_guard_score"], categories_df["count"])
        )
        print(f"Round {round_num} categories count: {categories_count}")
        print("-" * 50)
```
